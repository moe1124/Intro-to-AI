```python
# import numpy
import numpy as np
import matplotlib.pyplot as plt

# Make time variable
t = np.arange(0,2,0.01)

print (t)

# specify frequency of cosine
f = 1

# make cosine
y = np.cos(2*np.pi*f*t)
plt.plot(t,y)


```
