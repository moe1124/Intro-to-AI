```python
#import numpy
import numpy as np
import matplotlib.pyplot as plt

# make time variable
t = np.arange(0,2,0.01)

print(t)

# specify frequency of cosine
f = 1

#make cosine
y = np.cos(2*np.pi*f*t)
plt.plot(t,y,label = 'Original')

#print shape of t
print(t.shape[0])

#Add noise to y
z = np.random.normal(0,0.1,t.shape[0])
#plt.plot(t,z)
y_noisey = y+z
plt.plot(t,y_noisey,label = 'Noisy')

#import scipy
from scipy import signal

#make filter
b, a = signal.butter(5,0.1,'low')
y_filtered = signal.filtfilt(b,a,y_noisey)
plt.plot(t, y_filtered,label = 'Filtered')
plt.legend()
```
